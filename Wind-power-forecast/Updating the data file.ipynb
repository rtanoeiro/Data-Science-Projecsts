## We have to run this Python file every 7 days, as it will start from the last day of our dataset (7 days before) and get the data up until 7 days later (now)

## This scrpit needs to run every Friday 12:00PM UTC

```python
import pandas as pd
import datetime
import arrow
import requests
```

```python
for i in range(0,X): ##  Define X, up to 50, as data is limited for 50 requests per day
    
    data_tibau = pd.read_csv("data_tibau.csv", names=['time', 'velocity', 'direction'])
    data_tibau = data_tibau.sort_values(['time'])
    
    first_data = list(data_tibau['time'])[0]

    end = arrow.get(first_data) ## This will get the start date based on the last data of the previously created file
    start = end.shift(days=-9) ## Define end date

    lat, lng = -6.168371, -35.096696 # Tibau

    parameters=['windSpeed20m', 'windDirection20m']

    response = requests.get(
        'https://api.stormglass.io/v2/weather/point',
        params={
          'lat': lat,
          'lng': lng,
          'params': ','.join(parameters),
          'start': start,
          'end': end,
          'source':'sg'
        },
        headers={
            'Authorization': '##your authorization here##'
        }
    )

    # Do something with response data.
    json_data20_tibau = response.json()

    time_tibau = []
    velocity_tibau = []
    direction_tibau = []

    for i in range(len(json_data20_tibau['hours'])):
        time_tibau.append(json_data20_tibau['hours'][i]['time'])
        velocity_tibau.append(json_data20_tibau['hours'][i]['windSpeed20m'].get('sg'))
        direction_tibau.append(json_data20_tibau['hours'][i]['windDirection20m'].get('sg'))

    time_tibau = list(pd.to_datetime(time_tibau).strftime("%Y-%m-%d %H:%M:%S"))

    time_tibau.remove(time_tibau[-1])
    velocity_tibau.remove(velocity_tibau[-1])
    direction_tibau.remove(direction_tibau[-1])

    file = open('data_tibau.csv','a+') #Opening and writing the new information in our data file.,

    for i in range(len(time_tibau)):
        file.write('{}'.format(time_tibau[i]))
        file.write(",")
        file.write('{}'.format(velocity_tibau[i]))
        file.write(",")
        file.write('{}'.format(direction_tibau[i]))
        file.write('\n')  

    file.close()
    
    data_tibau = pd.read_csv("data_tibau.csv")
    data_tibau = data_tibau.sort_values(['time'])  ## Opening again to save in sorted way
    pd.to_csv("data_tibau.csv", index=False)
```
