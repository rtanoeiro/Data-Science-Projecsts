```python
import pandas as pd
import numpy as np
from datetime import datetime
import arrow
import requests
```

```python
start = arrow.get('201x-0x-xxT12:00:00+00:00') ## Change the xx to the date you want
end = start.shift(days=9)
start, end
```

```python
lat, lng = -6.168371, -35.096696 ## Remember to use the latitude and longitude of the place you want to check
```

```python
parameters=['windSpeed20m', 'windDirection20m'] ## Check other parameters at https://docs.stormglass.io/#/weather

response = requests.get(
    'https://api.stormglass.io/v2/weather/point',
    params={
      'lat': lat,
      'lng': lng,
      'params': ','.join(parameters),
      'start': start,
      'end': end,
      'source':'sg'
    },
    headers={
        'Authorization': 'your code'
    }
)

# Do something with response data.
json_data20 = response.json()
```

```python
time_tibau = []
velocity_tibau = []
direction_tibau = []

for i in range(len(json_data20['hours'])):
    time_tibau.append(json_data20['hours'][i]['time'])
    velocity_tibau.append(json_data20['hours'][i]['windSpeed20m'].get('sg'))
    direction_tibau.append(json_data20['hours'][i]['windDirection20m'].get('sg'))
    
time_tibau = pd.to_datetime(time_tibau).strftime("%Y-%m-%d %H:%M:%S")

## Saving the date into 3 arrays to save as a dataframe
```

```python
file = open('data_tibau.csv','w')

for i in range(len(time_tibau)):
    file.write('{}'.format(time_tibau[i]))
    file.write(",")
    file.write('{}'.format(velocity_tibau[i]))
    file.write(",")
    file.write('{}'.format(direction_tibau[i]))    
    file.write('\n')
    
file.close()
```
