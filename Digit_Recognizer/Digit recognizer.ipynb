## Deep Learning

```python
import numpy as np
import tensorflow as tf
```

```python
target_data = train_data['label']
train_data = train_data.drop("label",axis=1)
## Create our train and validation data for the deep learning algorith
```

```python
## Checking our data[
train_data
```

```python
## Importing relevant libraries

import keras
from keras import layers
```

```python
train_data_rescaled = train_data/255 ## Rescaled our data, as the values range from 0 to 255
```

```python
DL_model = keras.Sequential([keras.layers.Flatten(input_shape=(784,1)),
                          keras.layers.Dense(128, activation='sigmoid'),
                          keras.layers.Dense(128, activation='sigmoid'),
                          keras.layers.Dense(10, activation='softmax')])
```

```python
x_train, x_val, y_train, y_val =  tts(train_data_rescaled, target_data, test_size = 0.05)
```

```python
DL_model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history = DL_model.fit(x=x_train,
          epochs=12,
          y=y_train,
          validation_data=(x_val,y_val),verbose=2)
```

```python
DL_predictions = DL_model.predict(test_data)
```

```python
DL_predictions
```

```python
classes =  y_train.unique()
classes.sort
```

```python
results = []
import numpy as np

for i in DL_predictions:
    results.append(np.argmax(i))
```

```python
results
```

```python
from sklearn.metrics import classification_report
import numpy as np

results_val = []
val_predictions = DL_model.predict(x_val)

for i in val_predictions:
    results_val.append(np.argmax(i))
```

```python
print(classification_report(y_val,results_val,labels=classes))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_val,results_val, labels=classes)
```

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(1, len(acc) + 1)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1,2,figsize=(18,10))

ax[0].plot(epochs, acc, label='Training accuracy')
ax[0].plot(epochs,val_acc, label='Validation accuracy')
ax[0].set_title("Training and Validation Accuracy")
ax[0].legend()

ax[1].plot(epochs, loss, label='Training loss' )
ax[1].plot(epochs,val_loss, label='Validation loss')
ax[1].set_title("Training and Validation Loss")
ax[1].legend()
```
