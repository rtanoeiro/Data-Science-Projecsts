# Linear Regression

Initially. We'll use a Linear regression to try to fit the model, since this seems the reasonable approach for this case.

```python
## Importing libraries to process linear regression

from sklearn.linear_model import LinearRegression
```

```python
## Defining features

features = ['Cement', 'BFS', 'Fly_Ash', 'Water', 'Superplasticizer', 'Coarser_agg',
       'Fine_agg', 'Days']
targets = ['Comp_str']
```

```python
## Importing model and fitting the data

model = LinearRegression(fit_intercept=True) ## We assume that if all features are 0, the comp_str is 0,
                                             ##if there's no material, there are no concrete, thus, no compt_str
results = model.fit(df[features], df[targets])
```

```python
## Checking the R2 of the fitted data

print("The R^2 for this dataset in this Linear Regression is: ", results.score(df[features],df[targets]))
```

Now, we're going to remove outliers from the dataset:

By checking your describe() function we clearly see that there are outliers in our dataset. I'll only remove outliers from the Compressive strength column at first.

```python
to_remove = [] ## Creating the list to remove the rows over each iteration

for column in df.select_dtypes(include=np.number).columns.tolist(): ## Iterating over the list of columns
    
    print("We are now removing the outliers from column", column, "...")
    mean = np.mean(df[column])
    std = np.std(df[column]) ## Saving mean and std into variables
    print("It's mean is {} and Standard deviation is {}".format(mean, std))
 
    to_remove_plus = df.loc[df[column]>(mean + 2.5*std)] ## Removing ourliers at the right of the data
    to_remove_minus = df.loc[df[column]<(mean - 2.5*std)] ## Removing ourliers at the left of the data
    to_remove = to_remove_plus.append(to_remove_minus) ## Creating the list to remove from the previous lists

    df.drop(to_remove.index, axis=0, inplace=True) ## Effectivily removing the lines with outliers from the data
    print("A total of", to_remove[column].count(), "lines were  removed from this column\n") ## informing how many removed
```

```python
## Checking the R2 of the fitted data

print("The R^2 for this dataset in this Linear Regression is: ", results.score(df[features],df[targets]))
```
